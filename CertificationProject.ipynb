```python
import matplotlib.pyplot as plt
import seaborn as sns

funding_by_year = data.groupby('year')['amount'].sum().reset_index()

plt.figure(figsize=(10, 6,))
sns.lineplot(x='year', y='amount', data=funding_by_year, marker='o',linestyle="--")
plt.title("Year wise amount investment",color="red",fontsize=16)
plt.xlabel('Year')
plt.ylabel('Total Funding Amount')
plt.grid(True,linestyle="--")
plt.xticks(funding_by_year['year'])
plt.show()
```

```python
#which city have highest number of startups?
import matplotlib.pyplot as plt
import pandas as pd

demo = data['city'].value_counts().head(10)

plt.figure(figsize = (10,6))
color = ["blue","black","red","green","purple","orange","brown","pink","skyblue","yellow"]
plt.bar(demo.index,demo.values, color = color,alpha = 0.4, edgecolor = "black")
plt.xticks(rotation = 45)
plt.title("city wise startups", color = "red",fontsize = 20)
plt.xlabel("values", color = "red", fontsize = 20)
plt.ylabel("Cities", color = "red", fontsize = 20)
plt.grid(linestyle = "--", color = "pink")
plt.legend()
plt.show()
```

```python
#which city have highest number of startups?
import matplotlib.pyplot as plt
import pandas as pd

demo = data['vertical'].value_counts().head(10)

plt.figure(figsize = (10,6))
color = ["blue","black","red","green","purple","orange","brown","pink","skyblue","blue"]
plt.barh(demo.index[::-1],demo.values[::-1], color = color,alpha = 0.6, edgecolor = "black")
plt.xticks(rotation = 45)
plt.title("Top 10 Sectors in startup in India", color = "red",fontsize="20")
plt.xlabel("Number of startup comany", color = "red", fontsize = 20)
plt.ylabel("Sectors", color = "red", fontsize = 20)
plt.grid(axis='x',linestyle = "--", color = "grey")
plt.legend()
plt.show()
```
